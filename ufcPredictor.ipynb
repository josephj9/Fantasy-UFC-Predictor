```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
data = pd.read_csv("ufc-master.csv")
```

```python
data.head()
```

```python
cols = [
    "RedCurrentLoseStreak", "RedCurrentWinStreak", "RedLongestWinStreak",
    "BlueCurrentLoseStreak", "BlueCurrentWinStreak", "BlueLongestWinStreak",
    "RedLosses", "RedWins", "BlueLosses", "BlueWins",
    "RedTotalRoundsFought", "BlueTotalRoundsFought",
    "RedTotalTitleBouts", "BlueTotalTitleBouts",
    "RedWinsByKO", "RedWinsBySubmission", "RedWinsByDecisionUnanimous",
    "BlueWinsByKO", "BlueWinsBySubmission", "BlueWinsByDecisionUnanimous",
    "RedAvgSigStrLanded", "RedAvgSigStrPct",
    "RedAvgSubAtt", "RedAvgTDLanded", "RedAvgTDPct",
    "BlueAvgSigStrLanded", "BlueAvgSigStrPct",
    "BlueAvgSubAtt", "BlueAvgTDLanded", "BlueAvgTDPct",
    "RedHeightCms", "RedReachCms", "RedWeightLbs", "RedAge",
     "BlueHeightCms", "BlueReachCms", "BlueWeightLbs", "BlueAge", "RedOdds", 'BlueOdds', 'RedExpectedValue', "BlueExpectedValue",
    "RedDecOdds", "BlueDecOdds", "RSubOdds", "BSubOdds", "RKOOdds", "BKOOdds",
     "Winner"
]
```

```python
df = data[cols]
df= df.dropna()
df.head()
```

```python
df.to_csv("ufc_cleaned.csv", index=False)
```

```python
dataf = pd.read_csv("ufc_cleaned.csv")
```

```python
hashmap = {"Blue": 0, "Red": 1}
df["Winner"] = df["Winner"].map(hashmap)

X =  dataf.drop(columns=["Winner"])
y = df["Winner"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
X_train[0]
```

```python
model = keras.Sequential([ 
    keras.layers.Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    keras.layers.Dropout(0.4), 
    keras.layers.Dense(32, activation="relu"),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(16, activation="relu"),
    keras.layers.Dense(1, activation="sigmoid") 
])
model.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test))
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(round(accuracy * 100, 2), "%")
```

```python
model.save("fight_predictor.h5")
```
